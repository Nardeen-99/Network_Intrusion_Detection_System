```python
df = pd.read_csv("TON_IOT_pre.unknown")
```

```python
print(df.shape)
df.head()
```

```python
X = df.drop(['Label','Attack'],axis=1)
y = df['Label']

X_test = df.drop(['Label','Attack'],axis=1)
y_test = df['Label']
X=pd.get_dummies(X,drop_first=True)
X_test = pd.get_dummies(X_test,drop_first=True)
oversample = SMOTE()
```

```python
X, y = oversample.fit_resample(X, y)
label_encoder = preprocessing.LabelEncoder()
scaler=preprocessing.StandardScaler()
```

```python
X=scaler.fit_transform(X)
X_test=scaler.fit_transform(X_test)

y= label_encoder.fit_transform(y)
y_test = label_encoder.fit_transform(y_test)

```

```python
from numpy import mean
from numpy import std
# prepare the cross-validation procedure
cv = KFold(n_splits=5, random_state=1, shuffle=True)
# create model
model_2 = RandomForestClassifier(random_state = 1099)
# evaluate model
scores = cross_val_score(model_2, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
